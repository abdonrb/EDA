```python
resultados = pd.read_csv("LaLiga_Matches.csv")
barca = resultados[(resultados['HomeTeam'] == 'Barcelona') |( resultados['AwayTeam']== 'Barcelona')]
barca
```

```python
barca.rename(columns={'FTHG':'HomeGoals'},inplace=True)
barca.rename(columns={'FTAG':'AwayGoals'},inplace=True)
barca.rename(columns={'FTR':'Win_or_Draw'},inplace=True)
barca.rename(columns={'HTHG':'HalfTime_HomeGoals'},inplace=True)
barca.rename(columns={'HTAG':'HalfTime_AwayGoals'},inplace=True)
barca.rename(columns={'HTR':'Win_HalfTime'},inplace=True)
```

```python
barca.rename(columns={'Win':'Win_Or_Draw'},inplace=True)
```

```python
barca['HomeTeam'] = barca['HomeTeam'].astype(str)
barca['AwayTeam'] = barca['AwayTeam'].astype(str)
```

```python
barca
```

```python
barca = barca.reset_index(drop=True)
barca['Points'] = 0
barca.loc[(barca['HomeTeam'] == 'Barcelona') & (barca['Win_or_Draw'] == 'H'), 'Points'] = 3
barca.loc[(barca['AwayTeam'] == 'Barcelona') & (barca['Win_or_Draw'] == 'A'), 'Points'] = 3
barca.loc[(barca['Win_or_Draw'] == 'D'), 'Points'] = 1
```

```python
barca_xavi = barca[barca['Season'] == '2023-24']
barca_xavi
```
