```python
sleep = pd.read_csv('minuteSleep_merged.csv')
```

```python
sleep['date'] = pd.to_datetime(sleep['date'])
sleep
```

```python
sleep['night'] = sleep['date'] - pd.to_timedelta((sleep['date'].dt.hour < 6).astype(int), unit='d')
sleep['night'] = pd.to_datetime(sleep['night'])
sleep['night'] = sleep['night'].dt.date  # Extraer solo la fecha

# Agrupar por noche para contar los minutos por sesión de sueño
sleep['minutes'] = 1  # Cada registro representa un minuto
nightly_sleep = sleep.groupby(['night','Id'])['minutes'].sum().reset_index()

# Convertir minutos a horas
nightly_sleep['hours_asleep'] = nightly_sleep['minutes'] / 60
```

```python
nightly_sleep.to_csv('hours_sleep.csv')
```
